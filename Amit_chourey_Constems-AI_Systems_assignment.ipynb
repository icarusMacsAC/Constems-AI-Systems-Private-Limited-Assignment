## Question 1 : Design a function to generate random numbers without the help of any internal or external library of python(build the function from scratch)

```python
def randno(min_x, max_x, t = False):
    if min_x >= max_x: return "min can't grater then equal to max "
    if t == False:
        t = True
        max_x *= 10**8
    sleep_time = 1 / (10 ** 15)
    dt = datetime.now()
    r = ( dt.year * dt.month * dt.day * dt.hour * dt.minute * dt.second * dt.microsecond ) % max_x
    if (r / 10 ** 8) < min_x:
        time.sleep(sleep_time)
        return randno(min_x, max_x, t)
    time.sleep(sleep_time)
    return r / (10 ** 8)

min_x, max_x = map(int, input().split())
randno(min_x, max_x)
```

## Question 2 : Prove that the function is generating random numbers by plotting the generated points with any visualization library of python(for example matplotlib).

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def random_size(min_x, max_x, pt):
    lst1 = []
    for i in range(pt):
        lst1.append(randno(min_x, max_x))
    return np.array(lst1)

no_of_points = 1000
arr1 = random_size(0, 1, no_of_points)
arr2 = random_size(0, 1, no_of_points)
plt.scatter(arr1[:no_of_points], arr2[:no_of_points])
plt.show()
```

```python
def using_square_circle(r, no_of_points):
    lst1 = []
    ar_quare = no_of_points
    for i in range(no_of_points):
        d = ((r - arr1[i]) ** 2 + (r - arr2[i]) ** 2) ** (1 / 2)
        if d <= r:
            lst1.append(d)
    ar_circle = len(lst1)
    return (4 * ar_circle) / ar_quare
r = 0.5
print(f"Value of pi : {using_square_circle(r, no_of_points)}\nActual value : {np.pi}")
```

### 3.2 : IInd way

```python
def using_polygon(r, side):
    circum = r * np.cos(np.deg2rad(90 - (360 / side / 2))) * 2 * side
    return circum / 2 / r
```
